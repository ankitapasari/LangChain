# LangChain

## What is LangChain?

- LangChain is an open-source framework for developing applications powered by large language models (LLMs).
- Enables developers to build applications such as:
  - Chatbots and personal assistants
  - Text summarization and analysis
  - Q&A over documents or structured data
  - Code generation and understanding
  - Interaction with APIs

## LangChain Model I/O Framework

- Core Classes in LangChain:
  - LLMs:
    - Standard models for text completion (e.g., OpenAIâ€™s GPT models).
    - Work with simple text strings as inputs and outputs.
      
  - Chat Models:
    - Take a list of chat messages as input and return a chat message output.

## Working With Documents in LangChain

### Introduction to RAG (Retrieval-Augmented Generation):
- LangChain enhances Generative AI by integrating external data into language model responses, overcoming the model's knowledge cut-off. This process uses Retrieval-Augmented Generation (RAG), where external data is retrieved and passed to the model during generation.

image.png

### Key Components of RAG in LangChain:
- **Document Loaders**
  - Load documents from various sources (HTML, PDF, text, etc.)
  - Supports cloud storage and public websites
  - Over 100 Document Loaders available   
- **Text Splitters**
  - Divide large documents into manageable chunks
  - Handle context window limits in language models
- **Text Embedding Models**
  - Converts text into vector representations
  - Supports similarity search, text comparison, sentiment analysis
- **Vector Stores**
  - Stores and searches embedded data
  - Examples: Chroma, FAISS, LanceDB
- **Retrievers**
  - Interface for returning documents based on a query
  - Utilizes distance metrics like cosine similarity
